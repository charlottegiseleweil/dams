```python
# import required libraries
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy
from shapely import geometry

%matplotlib inline
```

```python
# set directory paths
path_to_charlie = '../../..'
INPUT_DIR = os.path.join(path_to_charlie, 'outputs/Detect_07-02_Ultralytics_full_06-27_validation_subset')
LABEL_DIR = os.path.join(path_to_charlie, 'data/Darknet-inputs_imagery_6-7_made_6-21/validation_set/labels')
```

```python
# collect labels (bbox coords) for images
label_filepaths = []
for image in detected_images:
    filepath = os.path.join(LABEL_DIR, image)
    filepath = filepath.replace('.png', '.txt')
    label_filepaths.append(filepath)
    
# test
for filepath in label_filepaths:
    print(filepath)
    
```

```python
# visualize

# list of detected images
detected_images = os.listdir(INPUT_DIR)

# open label file

fig = plt.figure(figsize=(12, 144))

for i in range(len(detected_images)):
    with open(label_filepaths[i], 'r') as label:
        data = label.readline()
        if len(data) > 0:
            norm_w = float(data[20:28])
            norm_h = float(data[29:])
            w = int(norm_w * 419)
            h = int(norm_h * 419)
        else: 
            w = 0
            h = 0

    bbox = shapely.geometry.box(209.5 - w / 2, 209.5 - h / 2, 209.5 + w / 2, 209.5 + h / 2)

    x,y = bbox.exterior.xy
    img = mpimg.imread(os.path.join(INPUT_DIR, detected_images[i]))
    
    ax = fig.add_subplot(16,1,i+1)
    ax.imshow(img)
    ax.plot(x,y)

plt.show()
```
